Ova tablica ne sadrži nedostajuće i stršeće podatke. <br>
Sljedeće ćemo vidjeti koje regije se koliko puta pojavljuju:

```python
columns = ['regionW', 'regionX', 'regionY', 'regionZ']
region_counts = {}

for column in columns:
    counts = s[column].value_counts().to_dict()
    for region, count in counts.items():
        region_counts[region] = region_counts.get(region, 0) + count

region_counts = sorted(region_counts.items(), key=lambda item: item[1], reverse=True)
for region, count in region_counts:
    print(f"{region}: {count}")
```

Iz sljedećeg prikaza možemo razlikovati dvije vrste regija: Geografski određene regije(East, West, Midwest, South, Southeast, Southwest)
koje se pojavljuju više puta <br>te regije koje su gradovi(Atlanta, Albuquerque, Phoenix, Chicago...)<br>
Imena tih gradova valjalo bi zamijeniti sa geografskom regijom kojoj ti gradovi pripadaju i ponoviti brojanje.

```python
replacements = {
    'Atlanta': 'South',
    'Albuquerque': 'West',
    'Phoenix': 'West',
    'Chicago': 'Midwest',
    'Oakland': 'West',
    'EastRutherford': 'East',
    'Austin': 'South',
    'Minneapolis': 'West',
    'StLouis': 'Midwest',
    'Syracuse': 'East',
    'WashingtonDC': 'East',
}
for column in columns:
    s[column] = s[column].replace(replacements)

region_counts = {}

for column in columns:
    counts = s[column].value_counts().to_dict()
    for region, count in counts.items():
        region_counts[region] = region_counts.get(region, 0) + count

region_counts = sorted(region_counts.items(), key=lambda item: item[1], reverse=True)
for region, count in region_counts:
    print(f"{region}: {count}")
```

#### Vizualizacija

Prvo ćemo prikazati broj pobjeda i poraza svake momčadi ikada

```python
games_won = rsdr['wteam'].value_counts().reset_index()
games_won.columns = ['Team_ID', 'Games_Won']

games_lost = rsdr['lteam'].value_counts().reset_index()
games_lost.columns = ['Team_ID', 'Games_Lost']

team_stats = pd.merge(games_won, games_lost, on='Team_ID', how='outer').fillna(0)

teams = team_stats['Team_ID']
games_won = team_stats['Games_Won']
games_lost = team_stats['Games_Lost']


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5), sharey=True)
ax1.bar(teams, games_won, color='green', label='won')
ax1.legend(loc='best')
ax1.set_xlabel('Team Id')
ax1.set_ylabel('Number of games')
ax1.grid()
ax2.bar(teams, games_lost, color='red', label='lost')
ax2.legend(loc='best')
ax2.set_xlabel('Team Id')
ax2.grid()
plt.show()
```

Želimo također vidjeti postotak pobjeda svake momčadi <br>
Tako da ćemo prvo prikazati postotak za svaku momčad i onda ćemo prikazati poredane postotke

```python
team_stats['Percentage_Won'] = (team_stats['Games_Won'] / (team_stats['Games_Won'] + team_stats['Games_Lost'])) * 100
percentage_won = team_stats['Percentage_Won']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5), sharey=True)
ax1.bar(teams, percentage_won, label='% Won')
ax1.legend(loc='best')
ax1.set_xlabel('Team Id')
ax1.set_ylabel('Percentage')
ax1.grid()

team_stats = team_stats.sort_values(by='Percentage_Won', ascending=False)
index = range(len(percentage_won))
ax2.bar(index, percentage_won, label='% Won')
ax2.legend(loc='best')
ax2.grid()

plt.show()
```

##### Razlika pobjednik-gubitnik

```python
plt.boxplot([rsdr.wscore - rsdr.lscore, rsdr.wfgm - rsdr.lfgm, rsdr.wfta - rsdr.lfta, rsdr.wfgm3 - rsdr.lfgm3, 
            rsdr.wfga3 - rsdr.lfga3, rsdr.wftm - rsdr.lftm, rsdr.wfta - rsdr.lfta, rsdr.wor - rsdr.lor,
            rsdr.wdr - rsdr.ldr, rsdr["wast"] - rsdr["last"], rsdr.wto - rsdr.lto, rsdr.wstl - rsdr.lstl, 
            rsdr.wblk - rsdr.lblk, rsdr.wpf - rsdr.lpf], 
            labels=["Score", "Fgm", "Fga", "Fgm3", "Fga3", "Ftm", "Fta", "Or", "Dr", "Ast", "To", "Stl", "Blk", "Pf"], showfliers=False)
plt.title('Razlike statistika pobjednika i gubitnika')
plt.grid()
plt.show()
```

Kod razlike statistika pobjednika gubitnika, zanimljivije nam je vidjeti koje razlike su pozitivne a koje negativne. <br>
Možemo zaključiti da pošto je razlika poena uvijek pozitivna, pošto pobjednik uvijek ima više poena nego gubitnik, da razlike <br>
koje su pozitivne u prosjeku više doprinose pobjedi te razlike koje su negativne više dobrinose porazu.

```python
diff = pd.DataFrame()
diff['score'] = rsdr['wscore'] - rsdr['lscore']
diff['fgm'] = rsdr['wfgm'] - rsdr['lfgm']
diff['fga'] = rsdr['wfga'] - rsdr['lfga']
diff['fgm3'] = rsdr['wfgm3'] - rsdr['lfgm3']
diff['fga3'] = rsdr['wfga3'] - rsdr['lfga3']
diff['ftm'] = rsdr['wftm'] - rsdr['lftm']
diff['fta'] = rsdr['wfta'] - rsdr['lfta']
diff['or'] = rsdr['wor'] - rsdr['lor']
diff['dr'] = rsdr['wdr'] - rsdr['ldr']
diff['ast'] = rsdr['wast'] - rsdr['last']
diff['to'] = rsdr['wto'] - rsdr['lto']
diff['stl'] = rsdr['wstl'] - rsdr['lstl']
diff['blk'] = rsdr['wblk'] - rsdr['lblk']
diff['pf'] = rsdr['wpf'] - rsdr['lpf']

plt.figure(figsize=(8, 6))
sns.heatmap(diff.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Korelacija razlika statistika pobjednika i gubitnika')
plt.show()
```

Ovdje možemo vidjeti zanimljivu ovisnost osobnih pogrešaka sa brojem zabijenih i pokušanih slobodnih bacanja <br>
Kako je ovdje korelacije između razlike pobjednika i gubitnika u tim značajkama negativna, možemo zaključiti da veći broj <br>
osobnih pogrešaka kod jednog tima uzrokuje veći broj slobodnih bacanja kod drugog tima
